```python
# Install Hugging Face datasets
!pip install datasets --quiet

# Load the full dataset
from datasets import load_dataset
import os

dataset = load_dataset("cats_vs_dogs", split="train")

# Set up directories in Google Drive
save_dir = "/content/drive/MyDrive/cats_vs_dogs"
os.makedirs(os.path.join(save_dir, "cats"), exist_ok=True)
os.makedirs(os.path.join(save_dir, "dogs"), exist_ok=True)

# Save all images with separate counters for cats and dogs
from tqdm import tqdm

cat_counter = 0
dog_counter = 0

for example in tqdm(dataset, desc="Saving images"):
    if example["labels"] == 0:
        filename = f"cats_{cat_counter}.jpg"
        path = os.path.join(save_dir, "cats", filename)
        cat_counter += 1
    else:
        filename = f"dogs_{dog_counter}.jpg"
        path = os.path.join(save_dir, "dogs", filename)
        dog_counter += 1
    example["image"].save(path)
```
